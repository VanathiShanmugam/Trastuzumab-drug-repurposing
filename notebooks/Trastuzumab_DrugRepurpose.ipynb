```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 6)
```

```python
try:
    df_X_drug = pd.read_csv("/content/feature_matrix_X.csv")
    df_Y = pd.read_csv("/content/label_matrix_Y_individual_drugs.csv")
except FileNotFoundError as e:
    print(f"Error: Required file not found: {e}. Cannot proceed.")
    exit()

df_aligned = pd.merge(df_X_drug, df_Y, on='Sample_ID', how='inner')

X_data = df_aligned.drop(columns=df_Y.columns.tolist())
Y_data = df_aligned[df_Y.columns.tolist()].drop(columns=['Sample_ID'])
drug_targets = Y_data.columns.tolist()
num_samples = len(X_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols_to_visualize_before = ['COL1A1', 'TGM2', 'JUP', 'SCN5A']

plt.figure(figsize=(15, 5))
for i, col in enumerate(cols_to_visualize_before):
    plt.subplot(1, len(cols_to_visualize_before), i + 1)
    sns.histplot(X_data[col], kde=True)
    plt.title(f'Distribution of {col} (Before Scaling)')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
cols_to_scale = X_data.select_dtypes(include=np.number).columns.drop(['ECM_Pathway_Score', 'Sarcomere_Mutation_Status'], errors='ignore')
X_to_scale = X_data[cols_to_scale]
X_auxiliary = X_data[['ECM_Pathway_Score', 'Sarcomere_Mutation_Status']]
scaler = StandardScaler()
X_scaled_array = scaler.fit_transform(X_to_scale)
X_scaled_df = pd.DataFrame(X_scaled_array, columns=X_to_scale.columns)
X_full = pd.concat([X_scaled_df, X_auxiliary.reset_index(drop=True)], axis=1)

print(f"Data Prepared. Total samples (N): {num_samples}. Features scaled.")
```

```python
results_list = []
drug_models = {}
X_train_dict = {}
X_test_dict = {}
Y_train_dict = {}
Y_test_dict = {}
model_type = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
test_size = 0.4

for drug in drug_targets:
    Y_target = Y_data[drug]
    drug_name = drug.replace('_Response', '')
    if Y_target.nunique() < 2 or Y_target.value_counts().min() < 2:
        results_list.append({'Drug': drug_name, 'AUC': 'N/A (Insufficient Data)', 'Accuracy': 'N/A', 'Precision': 'N/A', 'Recall': 'N/A', 'Top 5 Predictors': 'N/A'})
        continue

    X_train, X_test, Y_train, Y_test = train_test_split(
        X_full, Y_target, test_size=test_size, random_state=42, stratify=Y_target
    )
    X_train_dict[drug_name] = X_train
    X_test_dict[drug_name] = X_test
    Y_train_dict[drug_name] = Y_train
    Y_test_dict[drug_name] = Y_test
    model = model_type
    model.fit(X_train, Y_train)
    drug_models[drug_name] = model
    Y_pred_proba = model.predict_proba(X_test)[:, 1]
    Y_pred_class = model.predict(X_test)

    try:
        auc = roc_auc_score(Y_test, Y_pred_proba)
        accuracy = accuracy_score(Y_test, Y_pred_class)
        precision = precision_score(Y_test, Y_pred_class, zero_division=0)
        recall = recall_score(Y_test, Y_pred_class, zero_division=0)
    except ValueError:
        auc, accuracy, precision, recall = np.nan, np.nan, np.nan, np.nan

    importances = pd.Series(model.feature_importances_, index=X_full.columns)
    top_predictors = importances.nlargest(5).index.tolist()

    results_list.append({
        'Drug': drug_name,
        'AUC': f"{auc:.3f}" if not np.isnan(auc) else "Undetermined",
        'Accuracy': f"{accuracy:.3f}" if not np.isnan(accuracy) else "Undetermined",
        'Precision': f"{precision:.3f}" if not np.isnan(precision) else "Undetermined",
        'Recall': f"{recall:.3f}" if not np.isnan(recall) else "Undetermined",
        'Top 5 Predictors': ", ".join(top_predictors)
    })

df_results = pd.DataFrame(results_list)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve
from sklearn.model_selection import cross_val_score
import seaborn as sns
import matplotlib.pyplot as plt

evaluation_results = []

for drug_name, model in drug_models.items():
    if drug_name not in X_test_dict or len(X_test_dict[drug_name]) == 0:
        continue

    X_test = X_test_dict[drug_name]
    y_test = Y_test_dict[drug_name]
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else y_pred
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    auc = roc_auc_score(y_test, y_prob) if len(np.unique(y_test)) > 1 else np.nan
    evaluation_results.append({
        "Drug": drug_name,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1-Score": f1,
        "AUC": auc
    })
eval_df = pd.DataFrame(evaluation_results)

# Convert AUC to numeric for finding best drug and plotting
eval_df['AUC_numeric'] = pd.to_numeric(eval_df['AUC'], errors='coerce')

print("\n=== Model Evaluation Summary ===")
print(eval_df)

# Identify and print the best performing drug based on AUC
if 'eval_df' in locals() and not eval_df.empty and not eval_df['AUC_numeric'].dropna().empty:
    best_drug_row = eval_df.loc[eval_df['AUC_numeric'].idxmax()]
    best_drug_name = best_drug_row['Drug']
    best_drug_auc = best_drug_row['AUC_numeric'] # Use numeric AUC here
    print(f"\nBest performing drug based on AUC: {best_drug_name} (AUC: {best_drug_auc:.3f})")
else:
    print("\nCannot identify best performing drug: Evaluation results DataFrame (eval_df) not available or no numeric AUC values.")

print("\n=== Cross-validation Accuracy (5-fold) ===")
print("Skipping cross-validation due to small sample size and previous splitting approach.")


if 'eval_df' in locals() and not eval_df.empty:
    eval_df_plot = eval_df.dropna(subset=['AUC_numeric']) # Drop rows where AUC is not numeric

    if not eval_df_plot.empty:
        plt.figure(figsize=(10,6))
        sns.barplot(x='Drug', y='AUC_numeric', data=eval_df_plot.sort_values(by='AUC_numeric', ascending=False), palette='mako')
        plt.title("Model Performance (AUC per Drug)")
        plt.xticks(rotation=90)
        plt.show()
    else:
        print("\nCannot plot AUC summary: No numeric AUC values available in evaluation results.")
else:
    print("\nCannot plot AUC summary: Evaluation results DataFrame (eval_df) not available.")
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve
import plotly.graph_objects as go
from wordcloud import WordCloud

if 'best_drug_name' in locals() and 'Y_data' in locals() and 'X_full' in locals() and 'drug_models' in locals():
    print("\n" + "="*60)
    print(f"DETAILED EVALUATION FOR BEST DRUG: {best_drug_name}")
    print("="*60)

    # Assuming best_drug_name has been determined in the previous cell
    Y_best = Y_data[f"{best_drug_name}_Response"]

    # Check if the best drug response column has enough samples for splitting and stratification
    if Y_best.nunique() < 2 or Y_best.value_counts().min() < 2:
        print(f"Insufficient data for detailed evaluation of {best_drug_name}.")
    else:
        X_train, X_test, Y_train, Y_test = train_test_split(X_full, Y_best, test_size=0.4, random_state=42, stratify=Y_best)

        # Use the already trained model if available, otherwise train a new one
        if best_drug_name in drug_models:
            best_model = drug_models[best_drug_name]
        else:
            print(f"Model for {best_drug_name} not found. Training a new model for detailed evaluation.")
            best_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
            best_model.fit(X_train, Y_train)

        Y_pred = best_model.predict(X_test)
        Y_pred_proba = best_model.predict_proba(X_test)[:, 1]
        print(classification_report(Y_test, Y_pred, zero_division=0))

        cm = confusion_matrix(Y_test, Y_pred)
        plt.figure(figsize=(5,4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f'Confusion Matrix - {best_drug_name}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()

        fpr, tpr, _ = roc_curve(Y_test, Y_pred_proba)
        # Use the best_drug_auc calculated in the previous cell if available, otherwise recalculate
        if 'best_drug_auc' in locals():
             auc_score = best_drug_auc
        else:
             try:
                 auc_score = roc_auc_score(Y_test, Y_pred_proba)
             except ValueError:
                 auc_score = "Undetermined"

        plt.figure(figsize=(6,5))
        plt.plot(fpr, tpr, label=f"AUC = {auc_score:.3f}" if isinstance(auc_score, float) else f"AUC = {auc_score}")
        plt.plot([0,1], [0,1], linestyle='--', color='gray')
        plt.title(f'ROC Curve - {best_drug_name}')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.legend()
        plt.show()
else:
    print("Best performing drug information not available for detailed evaluation. Please run previous cells.")

print("\nGenerating Radar Chart for Model Comparison...")

if 'eval_df' in locals() and not eval_df.empty and not eval_df['AUC_numeric'].dropna().empty:
    df_numeric = eval_df.copy()
    df_norm = df_numeric[['Drug', 'AUC_numeric', 'Accuracy', 'Precision', 'Recall']].dropna()
    metrics = ['AUC_numeric', 'Accuracy', 'Precision', 'Recall']
    fig = go.Figure()
    for _, row in df_norm.iterrows():
        fig.add_trace(go.Scatterpolar(
            r=row[metrics].values,
            theta=['AUC', 'Accuracy', 'Precision', 'Recall'], # Use original metric names for theta
            fill='toself',
            name=row['Drug']
        ))

    fig.update_layout(
        polar=dict(radialaxis=dict(visible=True, range=[0, 1])),
        title='Radar Chart: Performance Comparison Across Drugs',
        showlegend=False
    )
    fig.show()
else:
    print("Cannot generate Radar Chart: Evaluation results DataFrame (eval_df) not available or no numeric AUC values.")


print("\nGenerating Word Cloud of Top Predictors...")
if 'df_results' in locals() and not df_results.empty and 'Top 5 Predictors' in df_results.columns:
    all_predictors = " ".join(df_results['Top 5 Predictors'].dropna())
    if all_predictors:
        wordcloud = WordCloud(width=800, height=400, background_color='black', colormap='plasma').generate(all_predictors)

        plt.figure(figsize=(10,5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('Word Cloud of Top Predictors Across All Drugs', fontsize=14)
        plt.show()
    else:
        print("No top predictors available to generate word cloud.")
else:
     print("Cannot generate Word Cloud: df_results DataFrame not available or 'Top 5 Predictors' column is missing.")
```
